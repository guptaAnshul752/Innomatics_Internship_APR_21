## Problem - 1:

```python
# Number of commands
N = int(input())
    
li = []
    
for _ in range(int(N)):
    
    arg = input().split(" ")
    
    if arg[0] == "insert":
        li.insert(int(arg[1]), int(arg[2]))
        
    elif arg[0] == "append":
        li.append(int(arg[1]))
    
    elif arg[0] == "remove":
        li.remove(int(arg[1]))
        
    elif arg[0] == "pop":
        li.pop()
        
    elif arg[0] == "sort":
        li.sort()
        
    elif arg[0] == "reverse":
        li.reverse()
        
    elif arg[0] == "print":
        print('List = {}'.format(li))
```

## Problem 4:

```python
list = []
    
for _ in range(int(input())):
    name = input()
    score = float(input())
            
    list.append([name,score])
            

scores = [i[1] for i in list]

min_score = min(scores)
    
for i in range(len(list)):
        
    if list[i][1] == min_score:
        scores.remove(min_score)
        
minScore = min(scores)

names = []

for i in range(len(list)):
    if list[i][1] == minScore:
        names.append(list[i][0])
                
names.sort()

print("Students having 2nd lowest grades are: ")
for i in names:
    print(i)
```

## Problem 14:

```python
N = int(input())

inp = input().split()

setA = set(map(int,inp))
    
commands = int(input())

for t in range(commands):
    
    args = input().strip().split(" ")
    
    setT = set(map(int,input().split()))
    
    if args[0] == "intersection_update":
        setA.intersection_update(setT)
        
    elif args[0] == "update":
        setA.update(setT)
        
    elif args[0] == "symmetric_difference_update":
        setA.symmetric_difference_update(setT)
        
    elif args[0] == "difference_update":
        setA.difference_update(setT)
        
print("Sum of elements in set A: " , sum(setA))
```
