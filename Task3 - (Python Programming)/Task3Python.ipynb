# Math Based Python Problems

## Problem1:

```python
import cmath

Num = complex(input())

print("Absolute Value: " ,abs(complex(Num.real,Num.imag)))

print("Phase Angle: " ,cmath.phase(complex(Num.real,Num.imag)))
```

## Problem 2:

```python
import math

AB = int(input("Enter side 1: "))

BC = int(input("Enter side 2: "))

hypo = math.hypot(AB,BC)

res=round(math.degrees(math.acos(BC/hypo))) 

degree=chr(176)        
                       
print("Angle MBC : " ,res,degree, sep='')
```

## Problem 3:

```python
N = int(input("Enter size: "))
print("Palindromic Triangle: ")
for i in range(1,N+1):
    print((10**(i)//9)**2)
```

## Problem 4:

```python
N = int(input("Enter 1st value: "))

M = int(input("Enter 2nd value: "))

print("Integer Division : " ,N//M)

print("Modulo Operator: " ,N%M)

print("Divmod :" ,divmod(N,M))
```

## Problem 5:

```python
A = int(input("Enter 1st number: "))

B = int(input("Enter 2nd number: "))

M = int(input("Enter 3rd number: "))

print("Power : ", pow(A,B))

print("Mod Power : " ,pow(A,B,M))
```
