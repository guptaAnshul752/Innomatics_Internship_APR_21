## Problem 3:

```python
import re

str = input()

res = re.match(r'.*?([a-z0-9A-Z])\1', str)

if res is None:
    print(str(-1))
else:
    print(res.group(1))
```

## Problem 5:

```python
import re

S = input().strip()

K = input().strip()

pattern = re.compile(f"(?={K})")

if not re.search(pattern,S):
    
    print((-1, -1))
    
else:
    for i in re.finditer(pattern, S):
        
        start, end = i.span()
        
        print((start, end+len(K)-1))
```

## Problem 8:

```python
import re

reg = r"<[a-z]+[\w\-.]*@[a-z]+\.[a-z]{1,3}>"

N = int(input())

for i in range(N):
    
    name,email=input().split()
    
    if re.match(reg,email,re.I):
        
        print(" ".join([name,email]))
```

## Problem 10:

```python
from html.parser import HTMLParser

class html(HTMLParser):
    
    def handle_starttag(self, tag, attrs):
        
        print("Start :", tag)
        
        for i in attrs:
            print("-> %s > %s"%(i))
            
    def handle_endtag(self, tag):
        
        print("End   :", tag)
        
    def handle_startendtag(self, tag, attrs):
        
        print("Empty :", tag)
        
        for attr in attrs:
            print("-> %s > %s"%(attr))


parser = html()

N = int(input())

for i in range(N):
    
    parser.feed(input())
```

## Problem 11:

```python
from html.parser import HTMLParser

class html(HTMLParser):
    
    def handle_comment(self, data):
        
        if '\n' in data:
            
            print('>>> Multi-line Comment')
            print(data)
            
        else:
            
            print('>>> Single-line Comment')
            print(data)

    def handle_data(self, data):
        
        if '\n' not in data:
            
            print(">>> Data")
            print(data)


code = '' 

N = int(input())

for i in range(N):
    
    code += input().rstrip()

    code += '\n'

parser = html()

parser.feed(code)

parser.close()
```

## Problem 12:

```python
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    
    def handle_starttag(self, tag, attrs):
        
        print(tag)
        
        for attr in attrs:
            
            print("->", attr[0], ">", attr[1])

parser = MyHTMLParser()

N = int(input())

for i in range(N):
    
    parser.feed(input())
```

## Problem 13:

```python
import re 

patts = [r'^[a-zA-Z0-9]{10}$', # Check proper contents/length
            r'(.*[A-Z].*){2,}',    # Count Upper Case
            r'(.*[0-9].*){3,}',    # Count Numbers
            r'^(?:([a-zA-Z0-9])(?!.*\1))*$'] # Match any string w/o repeating chars

reg = map(re.compile, patts)

N = int(input())

for i in range(N):
    
    Id = input().strip()
    
    print('Valid') if all([x.match(Id) for x in reg]) else print('Invalid') 
```
