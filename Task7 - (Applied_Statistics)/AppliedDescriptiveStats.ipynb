# Mean

- In statistics, Mean can be defined as a measure of central tendency of a probability distribution along median and mode. It is also referred to as an expected value.

- In simpler terms , Mean can be defined as the sum of values divided by the total number of values

- Mean is denoted by $\mu$ (mu)

    `Formula`
    
    $\mu$ = $\frac{1}{n} \sum_{i=i}^{n} x_{i}$ , where $n$ is the total number of values

```python
#Without Formula
Mean_Wf = sum(df['Mthly_HH_Income'])/len(df['Mthly_HH_Income'])

print('Mean without Formula: {}'.format(Mean_Wf))

#With Formula
Mean_f = np.mean(df['Mthly_HH_Income'])

print('Mean with Formula: {}'.format(Mean_f))
```

# Median

- The median is the middle value when a data set is ordered from least to greatest.

    `Formula`
    
    Case1: N is <b>Odd</b>
    
    $Median$ = $\frac{n}{2}$th value , where $n$ is the total number of values
    
    Case2: N is <b>Even</b>
    
    $Median$ = $avg$($\frac{n}{2}th$, $(\frac{n}{2} - 1)th$) value , where $n$ is the total number of values

```python
N = len(df['Mthly_HH_Income'])

print('Length:',N)

median_Wf = 0

vals = []

if N % 2 != 0:
    
    vals = sorted(df['Mthly_HH_Income'])
    median_Wf = df['Mthly_HH_Income'][(N+1)//2]
    
else:
    vals = sorted(df['Mthly_HH_Income'])
    median_Wf = (vals[N//2] + vals[(N//2)-1]) // 2

print('Median without Formula: {}'.format(median_Wf))


median_f = np.median(df['Mthly_HH_Income'])

print('Median with Formula: {}'.format(median_f))
```

# Mode

- The mode is the value that occurs the most in the set of values

```python
#Without Formula

vals = list(df['Mthly_HH_Income'].values)

vals.sort()

count = []

for i in range(len(vals)):
    
    count.append(vals.count(vals[i]))
    
# Dictionary to map the values with its corresponding count

d = dict(zip(vals, count))

#Stores maximum occuring characters (for multiple modes)
li = []

for i,j in d.items():
    if j == max(count):
        li.append(i)

print("Mode without formula : {}".format(li))


#With Formula
Mode_f = stats.mode(df['Mthly_HH_Income'])

print('Mode with Formula: {}'.format(Mode_f[0]))
```

# Variance

- The variance is a measure of variability. It is calculated by taking the average of squared deviations from the mean. 


- Variance tells the degree of spread in your data set. 


- Variance is denoted by $\sigma^2$

    `Formula`
    
  $\sigma^2$ = $\frac{\sum_{i=1}^{n}(x_i - \mu)^2}{n-1}$ , where $\mu$ is mean

```python
#Without Formula

vals = list(df['Mthly_HH_Income'].values)

#Mean
mean = sum(df['Mthly_HH_Income'])/len(df['Mthly_HH_Income'])

# For storing squared difference of values from its mean
sqterm = []

for i in range(len(vals)):
    
    sqterm.append((vals[i]-mean)**2)
    
var_wf = sum(sqterm) / len(sqterm)

print("Variance without formula : {}".format(var_wf))

#With Formula

var_f = np.var(df['Mthly_HH_Income'])

print("Variance with formula : {}".format(var_f))
```

# Standard Deviation

- A standard deviation is a statistic that measures the dispersion of a dataset relative to its mean 


- Variance is denoted by $\sigma$

    `Formula`
    
    $\sigma$ = $\sqrt{\sigma^2}$ , where $\sigma^2$ is the variance
 

```python
#Without Formula

vals = list(df['Mthly_HH_Income'].values)

#Mean
mean = sum(df['Mthly_HH_Income'])/len(df['Mthly_HH_Income'])

# For storing squared difference of values from its mean
sqterm = []

for i in range(len(vals)):
    
    sqterm.append((vals[i]-mean)**2)
    
var = sum(sqterm) / len(sqterm)

std_wf = var**0.5

print("Standard Deviation without formula : {}".format(round(std_wf,2)))

#With Formula

std_f = np.std(df['Mthly_HH_Income'])

print("Standard Deviation with formula : {}".format(round(std_f,2)))
```

# Correlation

- Correlation is a statistical measure that expresses the extent to which two variables are linearly related.

- It's a common tool for describing simple relationships without making a statement about cause and effect.

    `Formula`
    
 $\rho$ = $\frac{\text{cov}(X,Y)}{\sigma_x \sigma_y}$
 
   and estimate r

 $r$ = $\frac{{}\sum_{i=1}^{n} (x_i - \overline{x})(y_i - \overline{y})}
{\sqrt{\sum_{i=1}^{n} (x_i - \overline{x})^2(y_i - \overline{y})^2}}$

```python
# With Formula

# Returns Correlation Matrix
print('Correlation using formula:\n',np.corrcoef(df['Mthly_HH_Income'], df['Mthly_HH_Expense']))
print('\n')

# Using Scipy.stats
# Pearson's Correlation Coefficient

print("Pearson's Correlation Coefficient:",stats.pearsonr(df['Mthly_HH_Income'], df['Mthly_HH_Expense']))
```

# Normal Distribution

- Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.

`Formula`

$f(x)$ = $\frac{1}{\sigma\sqrt{2\pi}}$ $\exp\left( -\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^{\!2}\,\right)$

`Notation`

$X$ is normally distributed with mean $\mu$ and standard deviation $\sigma$:
$$X \sim \mathcal{N}(\mu,\,\sigma^{2})$$

### Properties of a normal distribution

- The mean, mode and median are all equal.


- The curve is symmetric at the center (i.e. around the mean, μ).


- Exactly half of the values are to the left of center and exactly half the values are to the right.


- The total area under the curve is 1.




<b>The Empirical Rule states that</b> , 

- 68% of the data falls within one standard deviation of the mean.


- 95% of the data falls within two standard deviations of the mean.


- 99.7% of the data falls within three standard deviations of the mean.

```python
plt.figure(figsize=(10,10))
plt.subplots_adjust(hspace=0.4, wspace=0.4)

plt.subplot(2,2,1)
sns.distplot(df['Mthly_HH_Income'])
plt.title('Mnthly_HH_Income')

plt.subplot(2,2,2)
sns.distplot(df['Mthly_HH_Expense'])
plt.title('Mthly_HH_Expense')

plt.subplot(2,2,3)
sns.distplot(df['No_of_Fly_Members'])
plt.title('No_of_Fly_Members')

plt.subplot(2,2,4)
sns.distplot(df['Annual_HH_Income'])
plt.title('Annual_HH_Income')

plt.show()
```

### Skewness

In statistics, skewness is a measure of the asymmetry of the probability distribution of a random variable about its mean. In other words, skewness tells you the amount and direction of skew (departure from horizontal symmetry). The skewness value can be positive or negative, or even undefined. If skewness is 0, the data are perfectly symmetrical, although it is quite unlikely for real-world data. As a general rule of thumb:

- If skewness is less than -1 or greater than 1, the distribution is highly skewed.


- If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.


- <b>If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.</b>

```python
print('Skewness of column {}: {}'.format('Mthly_HH_Income',df['Mthly_HH_Income'].skew()))
    
print('Skewness of column {}: {}'.format('Mthly_HH_Expense',df['Mthly_HH_Expense'].skew()))
    
print('Skewness of column {}: {}'.format('No_of_Fly_Members',df['No_of_Fly_Members'].skew()))
    
print('Skewness of column {}: {}'.format('Annual_HH_Income',df['Annual_HH_Income'].skew()))
```

```python
# Skewness of column No_of_Fly_members is between the standard range(-0.5,0.5) i.e it is more normally distributed(Symmetric)
```

### Kurtosis

Kurtosis tells you the height and sharpness of the central peak, relative to that of a standard bell curve.

```python
print('Kurtosis of column {}: {}'.format('Mthly_HH_Income',df['Mthly_HH_Income'].kurt()))
    
print('Kurtosis of column {}: {}'.format('Mthly_HH_Expense',df['Mthly_HH_Expense'].kurt()))
    
print('Kurtosis of column {}: {}'.format('No_of_Fly_Members',df['No_of_Fly_Members'].kurt()))
    
print('Kurtosis of column {}: {}'.format('Annual_HH_Income',df['Annual_HH_Income'].kurt()))
```

## Effect on Mean, Median and Mode due to Skewness

- If the distribution of data is skewed to the left, the mean is less than the median, which is often less than the mode. 


- If the distribution of data is skewed to the right, the mode is often less than the median, which is less than the mean.

# QQ Plot

- QQ Plot is used for checking the Normality of the distribution


- <b>For a Distribution to be Normal, Points should be along the 45 degree(red) line in the qq plot</b>


```python
plt.figure(figsize = (10,10))
plt.subplots_adjust(hspace=0.4, wspace=0.4)

plt.subplot(2,2,1)
stats.probplot(df['Mthly_HH_Income'] , dist = 'norm' , plot = plt)

plt.subplot(2,2,2)
stats.probplot(df['Mthly_HH_Expense'] , dist = 'norm' , plot = plt)

plt.subplot(2,2,3)
stats.probplot(df['No_of_Fly_Members'] , dist = 'norm' , plot = plt)

plt.subplot(2,2,4)
stats.probplot(df['Annual_HH_Income'] , dist = 'norm' , plot = plt)

plt.show()
```

# Box - Cox Transformation

- The Box-Cox transformation transforms our data so that it closely resembles a normal distribution

```python
x_t1, l1 = stats.boxcox(df['Mthly_HH_Income'])

x_t2, l2 = stats.boxcox(df['Mthly_HH_Expense'])

x_t3, l3 = stats.boxcox(df['No_of_Fly_Members'])

x_t4, l4 = stats.boxcox(df['Annual_HH_Income'])
```

```python
#After Box-Cox Transformation

plt.figure(figsize = (10,10))
plt.subplots_adjust(hspace=0.4, wspace=0.4)

plt.subplot(2,2,1)
stats.probplot(x_t1 , dist = 'norm' , plot = plt)

plt.subplot(2,2,2)
stats.probplot(x_t2 , dist = 'norm' , plot = plt)

plt.subplot(2,2,3)
stats.probplot(x_t3 , dist = 'norm' , plot = plt)

plt.subplot(2,2,4)
stats.probplot(x_t4 , dist = 'norm' , plot = plt)

plt.show()
```

# Pareto Distribution

- The Pareto distribution is a skewed distribution with heavy, or “slowly decaying” tails (i.e. much of the data is in the tails).

It has two main applications:

- To model the distribution of incomes.


- To model the distribution of city populations.

    `Formula`
    
 $f(x)$ = $1$ - ($\frac{k}{x})^\alpha$ , where $x$ is random variable , $k$ is lower bound of data , $\alpha$ is the shaping parameter

```python
pareto_rv = np.random.pareto(a = 2, size = 1000)

sns.distplot(pareto_rv)
```

## Log Normal Distibution

- A lognormal (log-normal or Galton) distribution is a probability distribution with a normally distributed logarithm. A random variable is lognormally distributed if its logarithm is normally distributed.


- Skewed distributions with low mean values, large variance, and all-positive values often fit this type of distribution. Values must be positive as log(x) exists only for positive values of x.

```python
lognormal_rv = np.random.lognormal(size = 1000)

sns.distplot(lognormal_rv)
```
