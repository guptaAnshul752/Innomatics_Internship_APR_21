# Research Question 1:

`Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh
graduate.”`

<b>Test this claim with the data given to you.</b>

#### Performing Hypothesis testing to validate the inference/claim with respect to given statement

# Hypothesis Test

First we make a bold claim against a statement

Step - 1:  Forming Null and Alternate Hypothesis

Alternate Hypothesis (Bold claim): $ H_1 \ \Rightarrow \ >, \ <, \ \neq  $  
Null Hypothesis (Status Quo): $ H_0 \ \Rightarrow \ \leq, \ \geq, \ = $

Step - 2:  Taking a sample and computing its mean

* Collect a sample of size n
* Compute the mean from this sample $ \bar{x} $

Step - 3: Compute Test Statistic: 

* If population variance is known $$ z = \frac{\bar{x} - \mu}{\sigma/\sqrt[2]{n}}$$

* If population variance is unknown $$ t = \frac{\bar{x} - \mu}{s/\sqrt[2]{n}} $$

Step - 4: Decide significance level $ \alpha $. Lower $ \alpha $ means you need stronger evidence to reject Null Hypothesis.

Step - 5.1: Apply decision rule:
* If test statistic is z-score -
    * Two tailed z-test: $$ |z| > z_\frac{\alpha}{2} \Rightarrow Accept \ H_1 \ or \ Reject \ H_0$$
    * Right tailed z-test: $$ z > z_\alpha \Rightarrow Accept \ H_1 \ or \ Reject \ H_0 $$
    * Left tailed z-test: $$ z < -z_\alpha \Rightarrow Accept \ H_1 \ or \ Reject \ H_0 $$
    
* If test statistic is t-score
    * Two tailed t-test: $$ |t| > t_{n-1,\frac{\alpha}{2}} \Rightarrow Accept \ H_1 \ or \ Reject \ H_0 $$
    * Right tailed t-test: $$ t > t_{n-1,\alpha} \Rightarrow Accept \ H_1 \ or \ Reject \ H_0 $$
    * Left tailed t-test: $$ t < t_{n-1,\alpha} \Rightarrow Accept \ H_1 \ or \ Reject \ H_0 $$

Step - 5.2: Compute p-value $ P(Test \ Statistics | H_0) $
* For two tailed test: $$ p \ value = 2 * (1.0 - cdf(test \ statistic)) $$
* For one tailed test: $$ p \ value = (1.0 - cdf(test \ statistic)) $$
Now, $$ if (p \ value < \alpha) \Rightarrow Accept \ H_1 \ or \ Reject \ H_0 $$

# Solution

First we make a bold claim against a statement

<b>Step - 1:</b>  Forming Null and Alternate Hypothesis

<b>Alternate Hypothesis (Bold claim) $ H_1$ :</b> On Average fresher can earn upto 300000 i.e $\mu$ $=$ $300000$

<b>Null Hypothesis (Status Quo) $ H_0 $ :</b> On Average fresher can earn greater than 300000 i.e $\mu$ $>$ $300000$

<b>Step - 2:</b>  Taking a sample and computing its mean

* Collect a sample of size n
* Compute the mean from this sample $ \bar{x} $

```python
# Taking sample from dataset
# Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer
```

```python
df.head()
```

```python
# Programmer Analyst
s1 = df[(df['Designation'] == 'programmer analyst') & (df['GraduationYear'] == 2014)].Salary.values

# Software Engineer
s2 = df[(df['Designation'] == 'software engineer') & (df['GraduationYear'] == 2014)].Salary.values

# Hardware Engineer
s3 = df[(df['Designation'] == 'hardware engineer') & (df['GraduationYear'] == 2014)].Salary.values

# Associate Engineer
s4 = df[(df['Designation'] == 'associate engineer') & (df['GraduationYear'] == 2014)].Salary.values
```

```python
sample = np.concatenate((s1,s2,s3,s4),axis=0)
```

```python
type(sample)
```

```python
len(sample)
```

```python
sample
```

```python
# Computing Sample Mean
# xbar
sample_mean = np.mean(sample).round(2)

print(sample_mean)
```

```python
# Population Mean
# Given
pop_mean = 300000
```


<b> Compute Test Statistic: </b>

* If population variance is unknown $$ t = \frac{\bar{x} - \mu}{s/\sqrt[2]{n}} $$


```python
# T-score for sampling distribution

def tscore(sample_mean,pop_mean,sam_std,sam_size):
    
    numerator = sample_mean - pop_mean
    denomenator = sam_std / sam_size**0.5
    return numerator / denomenator
```

```python
# Sample's Standard deviation

sq_terms = []

# Sample Size
N = len(sample)

def sample_std(sample):
    
    sample_mean = np.mean(sample).round(2)
    
    for i in range(N):
        
        sq_terms.append((sample[i] - sample_mean)**2)
        
    sampleStd = (sum(sq_terms) / (N-1))**0.5
    
    return sampleStd
```

```python
sampleStd = sample_std(sample).round(2)
sampleStd
```

```python
# Computing T-Score

t_score = tscore(sample_mean,pop_mean,sampleStd,N)

print(t_score)
```


<b> Decide significance level </b> $ \alpha $.

Lower $ \alpha $ means you need stronger evidence to reject Null Hypothesis.


```python
# One Tail - Calculating the t-critical value

confidence_level = 0.95

alpha = 1 - confidence_level

#Degree of freedom
df = N-1

t_critical = stats.t.ppf(1-alpha,df) # One tailed t score for 95% Confidence Level

print(t_critical)
```

<b>Step - 5.1:</b> Apply decision rule:
    
* If test statistic is t-score
    * Two tailed t-test: $$ |t| > t_{n-1,\frac{\alpha}{2}} \Rightarrow Accept \ H_1 \ or \ Reject \ H_0 $$
    * Right tailed t-test: $$ t > t_{n-1,\alpha} \Rightarrow Accept \ H_1 \ or \ Reject \ H_0 $$
    * Left tailed t-test: $$ t < t_{n-1,\alpha} \Rightarrow Accept \ H_1 \ or \ Reject \ H_0 $$


```python
# Two tailed test

if abs(t_score) > t_critical:
    print('Reject Null Hypothesis')
else:
    print('Fail to Reject Null Hypothesis')
```

<b>Step - 5.2:</b> Compute p-value $ P(Test \ Statistics | H_0) $
* For two tailed test: $$ p \ value = 2 * (1.0 - cdf(test \ statistic)) $$
* For one tailed test: $$ p \ value = (1.0 - cdf(test \ statistic)) $$
Now, $$ if (p \ value < \alpha) \Rightarrow Accept \ H_1 \ or \ Reject \ H_0 $$

```python
p_val = (1.0 - stats.t.cdf(t_score,df))

print("P Value = ", p_val)

if p_val < alpha:
    print('Reject Null Hypothesis')
else:
    print('Fail to Reject Null Hypothesis')
```

### Chi2 Test

Lets make a bold Claim that **Gender** and **Specialization** are dependent.

Step - 1:  
i.e. <b>Alternate Hypothesis:</b> $$ H_1: They \ are \ Dependent $$
and <b>Null Hypothesis:</b> $$ H_0: They \ are \ Independent $$

Step - 2:  
* Collect the sample of size n
* Compute the sample frequencies

```python
df.head()
```

```python
from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
# Step - 2 => Looking at the freqency distribution

# These are the observed frequencies

observed = pd.crosstab(df.Specialization, df.Gender, margins=True)

observed.columns = ['Female','Male','All']

observed
```

```python
# chi2_contigency returns chi2 test statistic, p-value, degree of freedoms, expected frequencies

chi2_contingency(observed)
```

Step - 3: Compute $ \chi^2 $ test statistic  
Now you need to check, if the difference in the observed and expected frequencies is too extreme to reject the NULL hypothesis.
* Have a look at Observed Frequencies (in the sample)
* Compute the Expected Frequencies (under null hyp assumption) $$ Expected \ Value = \frac{row \ total * col \ total}{grand \ total}$$
Now, test statistic can be computed using below mentioned formula: $$ \chi^2 = \Sigma \frac{(o-e)^2}{e} $$

Step - 4: Decide $ \alpha $ and $ df = (rows-1)(cols-1) $

Step - 5: Apply decision rule  
* Chi Square Test $$ if \ \chi^2 > \chi^2_{df, \alpha} \Rightarrow Reject H_0 $$
* p-value Test $$ p \ value = (1.0 - cdf(test \ statistic)) $$
Now, $$ if (p \ value < \alpha) \Rightarrow Accept H_1 \ or \ Reject H_0 $$

```python
# Computing chi2 test statistic, p-value, degree of freedoms

chi2_test_stat = chi2_contingency(observed)[0]

pval = chi2_contingency(observed)[1]

df = chi2_contingency(observed)[2]
```

```python
confidence_level = 0.90

alpha = 1 - confidence_level

chi2_critical = chi2.ppf(1 - alpha, df)

chi2_critical
```

```python
# Ploting the chi2 distribution to visualise

# Defining the x minimum and x maximum
x_min = 40
x_max = 140

# Ploting the graph and setting the x limits
x = np.linspace(x_min, x_max, 100)
y = chi2.pdf(x, df)
plt.xlim(x_min, x_max)
plt.plot(x, y)


# Setting Chi2 Critical value 
chi2_critical_right = chi2_critical

# Shading the right rejection region
x1 = np.linspace(chi2_critical_right, x_max, 100)
y1 = chi2.pdf(x1, df)
plt.fill_between(x1, y1, color='orange')
```

```python
if(chi2_test_stat > chi2_critical):
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
if(pval < alpha):
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

## Conclusion

Hence, from Chi- Sqaure test it can be derived that Gender & Specialization are not dependent on each other
